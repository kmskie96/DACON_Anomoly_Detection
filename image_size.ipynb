```python
from PIL import Image
```

```python
import glob
import matplotlib.pyplot as plt
```

```python
# 파일내 모든 사진 가져오기
train_path_list = glob.glob('data/train/*.png')
```

```python
train_path_list
```

```python
# train 이미지의 사이즈 저장
# pillow는 size => (width,height)
# opencv는 shape => (height,width,color)
img_size=[]
for path in train_path_list:
    img = Image.open(path)
    img_size.append(img.size)
    print(img_size)   
```

```python
# width와 height의 정보가 모두 같은지 알아보기
for size in img_size:
    try:
        if size[0] != size[1]:
            print('높이와 너비가 다릅니다!') 
            break
    except:
        pass

    # 모두 같은 것을 확인
```

```python
# 이미지 사이즈 줄여보기
image1 = Image.open(train_path_list[0])
image2 = image1.resize((750,750))
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(image1)
plt.subplot(122)
plt.imshow(image2)
```

## OpenCV를 이용한 크기조절
- 참고 : https://076923.github.io/posts/Python-opencv-8/

```python
import cv2
```

```python
img1 = cv2.imread(train_path_list[0],cv2.IMREAD_COLOR)
img1.shape
```

```python
# 이미지 축소
img2 = cv2.resize(img1,dsize=(200,200),interpolation=cv2.INTER_LINEAR)
img2.shape
```

```python
# BGR => RGB
img1_rgb = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img1_rgb)
plt.subplot(122)
plt.imshow(img2_rgb)
```
